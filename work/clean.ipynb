```python
import pandas as pd

df = pd.read_csv('data/data_cleaned.csv', encoding='utf-8')

# parse InvoiceDate into datetime, invalid parsing becomes NaT
df['date_commande'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

# drop rows where the parsed date is missing
df.dropna(subset=['date_commande'], inplace=True)

print(df.head())
```

```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Lecture du CSV avec encodage explicite
df = pd.read_csv(
    "data/data_cleaned.csv",
    encoding="latin1"  # ou "cp1252" si besoin
)

# 2. date_commande
df["date_commande"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

# 3. Connexion PostgreSQL
engine = create_engine(
    "postgresql+psycopg://admin:admin@localhost:5432/postgres"
)

# 4. Envoi dans PostgreSQL
df.to_sql("commandes1", engine, if_exists="replace", index=False)

print("IMPORT OK ðŸŽ‰")
```
